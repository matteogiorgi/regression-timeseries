```python
import pandas as pd
import numpy as np

data = pd.read_csv("/Users/guido/Desktop/project 2 time series modelling/arma12_seasonal_controlled_15series.csv")
```

```python
data["logseries_1"] = np.log(data["series_1"])
data["returnsseries_1"] = data["series_1"].pct_change().dropna()
data
```

```python
import matplotlib.pyplot as plt
plt.plot(data["series_1"])
plt.title("Series 1 Plot")
```

```python
plt.plot(data["logseries_1"])
plt.title("log of Series 1 plot")
```

```python
plt.plot(data["returnsseries_1"])
plt.title("plot of Series 1 returns")
```

```python
from statsmodels.tsa.stattools import adfuller
p = data["series_1"]
adftest = adfuller (p , maxlag = 21)
adftest
print("ADF statistic:", adftest[0])
print("p-value:", adftest[1])
print("Critical values:", adftest[4])
```

```python
from statsmodels.tsa.stattools import adfuller
returns = data["returnsseries_1"].dropna()
adftest2 = adfuller (returns , maxlag = 21)
adftest2
print("ADF statistic:", adftest2[0])
print("p-value:", adftest2[1])
print("Critical values:", adftest2[4])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(p, lags=20)
plt.savefig("acf1.png", dpi=300)
plt.show()

plot_pacf(p, lags=20)
plt.savefig("pacf1.png", dpi=300)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
model_01 = ARIMA(p, order=(1,0,1)).fit()
model_02 = ARIMA(p, order=(2,0,1)).fit()
model_03 = ARIMA(p, order=(2,0,2)).fit()
model_04 = ARIMA(p, order=(3,0,1)).fit()
model_05 = ARIMA(p, order=(3,0,2)).fit()
model_06 = ARIMA(p, order=(3,0,3)).fit()
model_07 = ARIMA(p, order=(1,0,3)).fit()
model_08 = ARIMA(p, order=(2,0,3)).fit()
model_09 = ARIMA(p, order = (1,0,2)).fit()
results = { (1, 1): model_01.aic, (2, 1): model_02.aic, (2, 2): model_03.aic,(3, 1): model_04.aic,(3, 2): model_05.aic, (3, 3): model_06.aic, (1,3): model_07.aic, (2,3): model_08.aic, (1,2):model_09.aic}
sortedresults = sorted(results.items(), key=lambda x: x[1])
print(sortedresults)
```

```python
final_model = ARIMA(p, order=(1,0,2), trend="n").fit()
print(final_model.summary())
```
